```python
# import cell
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
# import and inspect data
data = pd.read_csv("all_data.csv")
data.rename(columns={"Life expectancy at birth (years)":"life"},inplace= True)
data
```

### length of dataset

```python
data.Country.value_counts()
data.Year.value_counts()
len(data.life)
```

### countries spread by life expectancy

```python
plt.figure(figsize=(10,10))
sns.boxplot(data = data, x ="Country",y = "life")
```

## vizualizing the data

```python
sns.set_palette("Paired")
plt.figure(figsize= (20,20))
a=1
hmm = data["Country"].unique()
for country in hmm:
    plt.subplot(2,3,a)
    data1 = data[data.Country == country]
    sns.kdeplot(data = data1, x ="life")
    plt.title(country)
    plt.xlabel(country +" life expectancy")
    a+=1
    
plt.show()
# hmm = data.Country.unique().tolist()
```

```python
plt.figure(figsize= (20,20))
a=1
hmm = data["Country"].unique()
for country in hmm:
    plt.subplot(2,3,a)
    data1 = data[data.Country == country]
    plt.plot( data1.Year,data1.life)
    plt.title(country)
    plt.xlabel(country +" life expectancy")
    a+=1
    
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data = data, x ="Country",y = "GDP")
plt.title("Gdp by country")
```

```python
plt.figure(figsize= (20,20))
a=1
hmm = data["Country"].unique()
for country in hmm:
    plt.subplot(3,2,a)
    data1 = data[data.Country == country]
    plt.plot( data1.Year,data1.GDP)
    plt.title(country +" Gdp over the years")
    plt.xlabel("years")
    plt.ylabel("Gdp")
    a+=1
    
plt.show()
```

```python
plt.figure(figsize= (20,20))
a=1
hmm = data["Country"].unique()
for country in hmm:
    plt.subplot(3,2,a)
    data1 = data[data.Country == country]
    sns.scatterplot( data = data1, x =data1.life,y = data1.GDP)
    plt.title(country +" Gdp vs life")
    plt.xlabel("life expectancy")
    plt.ylabel("Gdp")
    a+=1
    
plt.show()
```
